```julia
#=s=6
T=5
t=Vector{Int64}(undef,s)
X=Vector{Matrix{Float64}}(undef,s)
eta=Vector{Float64}(undef,s)
func(a)=sin(a)+0.2*cos(5*a)#*(2*rand().-1)

for r=1:s
    t[r]=T
    X[r] = zeros(Float64,t[r],1)
    for j in 1:t[r]
        X[r][j,1] = 0.5*rand()-1.5+0.5*(r-1)
    end
    eta[r]=func(-1.5+0.5*(r-1))
end=#
```

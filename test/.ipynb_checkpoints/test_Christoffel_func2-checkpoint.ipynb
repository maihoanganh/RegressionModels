```julia
N=2
t=20
Y=Matrix{Float64}(undef,t,N)
randx=zeros(Float64,N)
for j=1:t
    randx=2*rand(Float64,N).-1
    randx=0.5*rand(Float64)*randx./sqrt(sum(randx.^2))
    Y[j,:]=randx-[0.25;0]
end
```

```julia
d=20
R=1
binomial(d+N,N)
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

Lambda=DensityEstimation.christoffel_func(N,Y,t,d)
#Lambda_Lebesgue=DensityEstimation.christoffel_func_Lebesgue(N,R,d)
```

```julia
using Plots

import Contour: contours, levels, level, lines, coordinates

function p_CF(x1,x2)
    if 1-x1^2-x2^2>0
        return d*Lambda([x1;x2])/(pi*sqrt(1-x1^2-x2^2))#/Lambda_Lebesgue([x1;x2])
    else
        return 0
    end

x1s = -1:0.01:1
x2s = -1:0.01:1

contour(x1s, x2s, p_CF,fill=true,title = "CF")

scatter!(Y[:,1], Y[:,2],label ="sample")
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation


x=DensityEstimation.solve_opt(N,Y,t,R,d,delta=1,s=1,rho=1,numiter=1e5,eps=1e-4,tol_eig=1e-3,ball_cons=true);
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

eval_PDF=DensityEstimation.func_eval_PDF(x,N,d,R,ball_cons=true)
```

```julia
using Plots

import Contour: contours, levels, level, lines, coordinates

p_approx(x1,x2)=eval_PDF([x1;x2])

x1s = -1:0.01:1
x2s = -1:0.01:1

contour(x1s, x2s, p_approx,fill=true,title = "MLE")

scatter!(Y[:,1], Y[:,2],label ="sample")
```

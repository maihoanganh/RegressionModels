```julia
using CSV
using DataFrames

df = CSV.read("../wine.csv", DataFrame)
```

```julia
nr,nc=size(df)
D=Matrix{Float64}(undef,nr,nc)
for j=1:nr
    for i=1:14
        D[j,i]=df[j,i]
    end
end
D
```

```julia
max_col=[maximum(D[:,j]) for j=2:14]
```

```julia
for j=2:14
    D[:,j]/=max_col[j-1]
end
D
```

```julia
D[:,2:14].-=0.5
D[:,2:14]*=2
```

```julia
using LinearAlgebra

max_norm_col=maximum(norm(D[j,2:14]) for j=1:nr) 
```

```julia
D[:,2:14]/=max_norm_col
D
```

```julia
Y1=D[findall(u -> u == 1, D[:,1]),2:14]
N=13
Y1
```

```julia
Y2=D[findall(u -> u == 2, D[:,1]),2:14]
```

```julia
Y3=D[findall(u -> u == 3, D[:,1]),2:14]
```

```julia
t1=ceil(Int64,0.8*size(Y1,1))
Y_train1=Y1[1:t1,:]
```

```julia
t2=ceil(Int64,0.8*size(Y2,1))
Y_train2=Y2[1:t2,:]
```

```julia
t3=ceil(Int64,0.8*size(Y3,1))
Y_train3=Y3[1:t3,:]
```

```julia
Y_test1=Y1[(t1+1):end,:]
```

```julia
Y_test2=Y2[(t2+1):end,:]
```

```julia
Y_test3=Y3[(t3+1):end,:]
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

d1=1
R1=1

x1=DensityEstimation.solve_opt(N,Y_train1,t1,R1,d1;delta=1,s=1,rho=1,
                         numiter=1e5,eps=1e-3,tol_eig=1e-3,ball_cons=true);
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

d2=2
R2=1

x2=DensityEstimation.solve_opt(N,Y_train2,t2,R2,d2;delta=1,s=1,rho=1,
                         numiter=1e5,eps=1e-3,tol_eig=1e-3,ball_cons=true);
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

d3=1
R3=1

x3=DensityEstimation.solve_opt(N,Y_train3,t3,R3,d3;delta=1,s=1,rho=1,
                         numiter=1e5,eps=1e-3,tol_eig=1e-3,ball_cons=true);
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

eval_PDF1=DensityEstimation.func_eval_PDF(x1,N,d1,R1,ball_cons=true)
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

eval_PDF2=DensityEstimation.func_eval_PDF(x2,N,d2,R2,ball_cons=true)
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

eval_PDF3=DensityEstimation.func_eval_PDF(x3,N,d3,R3,ball_cons=true)
```

```julia
function classifier(y)
    return findmax([eval_PDF1(y);eval_PDF2(y);eval_PDF3(y)])[2]
end
```

```julia
predict1=[classifier(Y_test1[j,:]) for j in 1:size(Y_test1,1)]
```

```julia
numcor1=length(findall(u -> u == 1, predict1))
```

```julia
predict2=[classifier(Y_test2[j,:]) for j in 1:size(Y_test2,1)]
```

```julia
numcor2=length(findall(u -> u == 2, predict2))
```

```julia
predict3=[classifier(Y_test3[j,:]) for j in 1:size(Y_test3,1)]
```

```julia
numcor3=length(findall(u -> u == 3, predict3))
```

```julia
accuracy=(numcor1+numcor2+numcor3)/(size(Y_test1,1)+size(Y_test2,1)+size(Y_test3,1))
```

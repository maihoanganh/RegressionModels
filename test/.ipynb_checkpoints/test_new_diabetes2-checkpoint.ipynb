```julia
using CSV
using DataFrames

df = CSV.read("../diabetes.csv", DataFrame)
```

```julia
nr,nc=size(df)
D=Matrix{Float64}(undef,nr,nc)
for j=1:nr
    for i=1:8
        D[j,i]=df[j,i]
    end
        
    if df[j,9]=="tested_negative"
        D[j,9]=0
    else
        D[j,9]=1
    end
end
D
```

```julia
max_col=[maximum(D[:,j]) for j=1:8]
```

```julia
for j=1:8
    D[:,j]/=max_col[j]
end
D
```

```julia
D[:,1:8].-=0.5
D[:,1:8]*=2
```

```julia
using LinearAlgebra

max_norm_col=maximum(norm(D[j,1:8]) for j=1:nr) 
```

```julia
D[:,1:8]/=max_norm_col
D
```

```julia
Y1=D[findall(u -> u == 0, D[:,end]),1:8]
N=8
Y1
```

```julia
Y2=D[findall(u -> u == 1, D[:,end]),1:8]
```

```julia
t1=ceil(Int64,0.8*size(Y1,1))
Y_train1=Y1[1:t1,:]
```

```julia
t2=ceil(Int64,0.8*size(Y2,1))
Y_train2=Y2[1:t2,:]
```

```julia
Y_test1=Y1[(t1+1):end,:]
```

```julia
Y_test2=Y2[(t2+1):end,:]
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

d1=1
R1=1

x1=DensityEstimation.solve_opt(N,Y_train1,t1,R1,d1;delta=1,s=1,rho=1,
                         numiter=1e5,eps=1e-3,tol_eig=1e-3,ball_cons=true);
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

d2=2
R2=1

x2=DensityEstimation.solve_opt(N,Y_train2,t2,R2,d2;delta=1,s=1,rho=1,
                         numiter=1e5,eps=1e-3,tol_eig=1e-3,ball_cons=true);
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

eval_PDF1=DensityEstimation.func_eval_PDF(x1,N,d,R)
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

eval_PDF2=DensityEstimation.func_eval_PDF(x2,N,d,R)
```

```julia
function classifier(y)
    if eval_PDF1(y)>eval_PDF2(y)
        return 1
    else
        return 2
    end
end
```

```julia
predict1=[classifier(Y_test1[j,:]) for j in 1:size(Y_test1,1)]
```

```julia
numcor1=length(findall(u -> u == 1, predict1))
```

```julia
predict2=[classifier(Y_test2[j,:]) for j in 1:size(Y_test2,1)]
```

```julia
numcor2=length(findall(u -> u == 2, predict2))
```

```julia
accuracy=(numcor1+numcor2)/(size(Y_test1,1)+size(Y_test2,1))
```

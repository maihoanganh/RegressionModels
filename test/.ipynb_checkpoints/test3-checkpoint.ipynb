```julia
using CSV
using DataFrames

df = CSV.read("../Dataset_of_Diabetes.csv", DataFrame)
```

```julia
nr,nc=size(df)
D=Matrix{Float64}(undef,nr,nc-2)
for j=1:nr
    for i=4:13
        D[j,i-2]=df[j,i]
    end
        
    if df[j,3]=="F"
        D[j,1]=0
    else
        D[j,1]=1
    end

    if df[j,14]=="N"
        D[j,12]=1
    elseif df[j,14]=="P"
        D[j,12]=2
    else
        D[j,12]=3
    end 
end
D
```

```julia
max_col=[maximum(D[:,j]) for j=1:11]
```

```julia
for j=1:11
    D[:,j]/=max_col[j]
end
D
```

```julia
D[:,1:11].-=0.5
D[:,1:11]*=2
```

```julia
using LinearAlgebra

max_norm_col=maximum(norm(D[j,1:11]) for j=1:1000) 
```

```julia
D[:,1:11]/=max_norm_col
D
```

```julia
Y=D[findall(u -> u == 1, D[:,end]),1:11]
N=size(Y,2)
```

```julia
t=ceil(Int64,0.9*size(Y,1))
Y_train=Y[1:t,:]
```

```julia
Y_test=Y[(t+1):end,:]
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

d=3
R=1

x=DensityEstimation.solve_opt(N,Y_train,t,R,d;delta=1,s=1,rho=1,
                         numiter=1e5,eps=1e-3,tol_eig=1e-3,ball_cons=true);
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

eval_PDF=DensityEstimation.func_eval_PDF(x,N,d,R)
```

```julia
[eval_PDF(Y_test[j,:]) for j in 1:size(Y_test,1)]
```

```julia
[eval_PDF(D[j,1:11]) for j in findall(u -> u == 2, D[:,end])]
```

```julia
N=1
t=50
Y=Matrix{Float64}(undef,t,N)
randx=zeros(Float64,N)
for j=1:t
    randx=2*rand(Float64,N).-1
    randx=rand(Float64)*randx./sqrt(sum(randx.^2))
    #randx=rand(Float64,N).-0.25
    Y[j,:]=randx
end
```

```julia
using CalculusWithJulia
using Plots
using SpecialFunctions
r=1
p_exact(z)=0.5
scatter(Y[:,1], zeros(Float64,t),label ="sample")#,aspect_ratio=:equal)
plot!(p_exact, -r, r, label ="exact")
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

d=10
R=1

Lambda=DensityEstimation.christoffel_func(N,Y,t,d)
#Lambda_Lebesgue=DensityEstimation.christoffel_func_Lebesgue(N,R,d)
```

```julia
p_CF(z)=2d*Lambda([z;zeros(N-1)])/(pi*sqrt(1-z^2))#/Lambda_Lebesgue([z;zeros(N-1)])
plot!(p_CF, -r, r, label = "CF",#=legend=:bottomright,=#title = "d = $(d)")
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

x=DensityEstimation.solve_opt(N,Y,t,R,d,delta=1,s=1,rho=1,numiter=1e5,eps=1e-4,tol_eig=1e-3,ball_cons=true);
```

```julia
include("../src/DensityEstimation.jl")
using .DensityEstimation

eval_PDF=DensityEstimation.func_eval_PDF(x,N,d,R,ball_cons=true)
```

```julia
p_approx(z)=eval_PDF([z;zeros(N-1)])
plot!(p_approx, -r, r, label = "MLE")
```
